```python
#Creating Average Exchange Rate list  :- Reference(bank bazaar credit to you website)
Rates = [47.19,48.61,46.58,45.32,44.1,45.31,41.35,43.51,48.41,
         45.73,46.67,53.44,56.57,60.50,62.97,66.46,67.79]
```

```python
#Creating a function to Convert FDI's value from USD to INR
def multiply_columns(df, col_list, num):
    for col in col_list:
        df[col] = df[col] * Rates[col_list.index(col)]/10
    return df
```

```python
df_InUSD=df.copy()
df_InINR = multiply_columns(df, Year, Rates)
```

```python
#FDI INFLOWS (Amount in ₹ Crores)
df_InINR.style.set_caption("FDI INFLOWS (Amount in ₹ Crores)").format(precision=2)
```

```python
# Unpivoting melt Dataframe
melt = pd.melt(df_InUSD, id_vars = Sectors, value_vars = Year, var_name='Year',
    value_name='FDI(US$ Million)',ignore_index=True)
melt
```

```python
#Unpivoting melt01 Dataframe
melt01 = pd.melt(df_InINR, id_vars = Sectors, value_vars = Year, var_name='Year',
    value_name='FDI(₹ Crores)',ignore_index=True)
melt01=round(melt01,2)
melt01
```

```python
# Merging the FDI(US$ Million) column of melt Dataframe into melto1 Dataframe
Merged=melt01.merge(melt,how='left')
Merged
```

```python
#Sorting the Sectors and Year columns
Sorted = Merged.sort_values(['Sector','Year'], ignore_index=True)
Sorted
```
